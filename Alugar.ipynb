```python
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('sao-paulo-properties-april-2019.csv')
```

```python
df.head()
```

```python
df_rent=df[df['Negotiation Type']=='rent']
df_rent['m2']=np.round(df_rent['Price']+df_rent['Condo']/df_rent['Size'],2)
df_rent
```

```python
df_rent['Price'].value_counts()
```

```python
df_rent.info()
```

```python
novo_rent=df_rent.drop(['Price','New','District','Property Type','Negotiation Type','Latitude','Longitude'],axis=1).values
y_rent=df_rent['Price']
```

```python
scaler = MinMaxScaler()
atributo_rent=scaler.fit(novo_rent)
rent_x=scaler.transform(novo_rent)
rent_x
```

```python
X_train, X_test, y_train, y_test = train_test_split(rent_x,y_rent,train_size=0.7)
X_train.shape
```

```python
gradient=GradientBoostingRegressor()
gradient.fit(X_train,y_train)
pred=gradient.predict(X_test)
r2_score(y_test,pred)
```

```python
pickle.dump(gradient, open('ModeloRent_GradientBoostingRegressor.pkl', 'wb'))
```
